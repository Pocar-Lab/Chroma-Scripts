```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
import pandas as pd
```

```python
import re
def parse_css_to_array(css_file):
    with open(css_file, "r") as f:
        lines = f.readlines()
    
    shape_line = lines[0]
    shape = tuple(map(int, re.findall(r'\d+', shape_line)))
    
    data = []

    for line in lines[1:]:
        if "--data" in line:
            values = line.split(':')[1].replace(';', '').strip().split(', ')
            values = [float(v) for v in values]
            data.append(values)

    
   
    # Reshape the data into the original array shape
    tracks = np.array(data).reshape(shape)
    return tracks
```
