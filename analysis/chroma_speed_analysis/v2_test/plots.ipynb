```python
import pandas as pd

# Load the CSV data
df = pd.read_csv("simulation_timing_results.csv")

# Compute the mean for each num_particles group
averaged_df = df.groupby("num_particles", as_index=False).mean()

# Save the averaged data to a new CSV file
averaged_df.to_csv("averaged_simulation_timing.csv", index=False)

print("Averaged results saved to averaged_simulation_timing.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the averaged CSV file
df = pd.read_csv("averaged_simulation_timing.csv")

# Take the log of num_particles for the x-axis
df["log_num_particles"] = np.log10(df["num_particles"])

# Plot the data
plt.figure(figsize=(8, 6))

plt.plot(df["log_num_particles"], df["geometry_manager_time"], label="Geometry Manager Time", marker="o")
plt.plot(df["log_num_particles"], df["run_manager_time"], label="Run Manager Time", marker="s")
plt.plot(df["log_num_particles"], df["total_time"], label="Total Time", marker="^")

# Labels & Titles
plt.xlabel("log(Number of Particles)")
plt.ylabel("Time (seconds)")
plt.title("Simulation Timing vs. Number of Particles")
plt.legend()
plt.grid(True)


# Plot the data
plt.figure(figsize=(8, 6))

plt.plot(df["num_particles"], df["geometry_manager_time"], label="Geometry Manager Time", marker="o")
plt.plot(df["num_particles"], df["run_manager_time"], label="Run Manager Time", marker="s")
plt.plot(df["num_particles"], df["total_time"], label="Total Time", marker="^")

# Labels & Titles
plt.xlabel("Number of Particles)")
plt.ylabel("Time (seconds)")
plt.title("Simulation Timing vs. Number of Particles")
plt.legend()
plt.grid(True)
```
